```python
import cv2
import os

#폴더 만드는 함수
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

# 프레인 단위로 쪼갤 영성
filename = 'v28'

#영상의 위치
vidcap = cv2.VideoCapture(f'./video/{filename}.mp4')

#filename과 같은 이름의 폴더가 없으면 생성
createFolder(filename)

count = 0
num =0

while(vidcap.isOpened()):
    ret, image = vidcap.read()
    if ret is False:
        break;
    imageHeight, imageWidth = image.shape[:2]
    
    #5분의1로 resize
    resizeHeight = int(0.2*imageHeight)
    resizeWidth = int(0.2*imageWidth)
    if not ret:
        break
    
    # 캡쳐된 이미지를 저장하는 함수 
    if count%2==0:  #숫자를 조절(움직임이 많은 영상은 작게, 움직임이 적으면 크게)

        #하다보면 가로영상이 세로로 압축된 경우 resizeWidth와 resizeHeight를 바꾼다.
        image = cv2.resize(image,(resizeWidth,resizeHeight), interpolation=cv2.INTER_CUBIC)
        cv2.imwrite(f'./{filename}/{filename}_{num}.jpg',image)
 
        print(f'Saved {filename}_{num}.jpg')
        num+=1
    count += 1
 
vidcap.release()
cv2.destroyAllWindows()
```

```python
# 바운딩 박스 크기 * 5
# 파일 읽어오고 오는대로 * 5 해서 문자열 + 로 저장 후, 내용 덮어씌우기
import os

video_name='v17'
index=0

while True:
    file_name=f'./test/{video_name}/{video_name}_{index}.txt'
    
    if os.path.isfile(file_name):
        f=open(file_name)
        for line in f.read().split('\n'):
            text = ''
            for word in line.split(' '):
                if word=='':
                    break
                if text == '':
                    text = word
                else: #if word > 0 and word < 1:
                    print(int(float(word)*5))
                    text = text + ' ' + str(float(word)*5)
                #print(text)
                
            fw = open(f'./test/{video_name}/new_{video_name}_{index}.txt', 'w')
            fw.write(text)
            fw.close()
            
            index+=1
    elif index > 361:
        break
    else:
        #print(index)
        index+=1
        continue
```

#### 3) 종류별 count

```python
import os

video_name='v28'
index=0

deer=0
beam=0
gcooter=0
others=0
while True:
    file_name=f'./{video_name}/{video_name}_{index}.txt'

    if os.path.isfile(file_name):
        f=open(file_name)
        for line in f.read().split('\n'):
            line=line.split(' ')
            if line[0]=='':
                break
            if int(line[0])==0:
                deer+=1
            elif int(line[0])==1:
                beam+=1
            elif int(line[0])==2:
                gcooter+=1
            elif int(line[0])==3:
                others+=1
        index+=1
    elif index > 361:
        break
    else:
        #print(index)
        index+=1
        continue

print(f'deer : {deer}개, beam : {beam}개, gcooter : {gcooter}개, others : {others}개')
print(f'총 사진: {index-1}개')
```
